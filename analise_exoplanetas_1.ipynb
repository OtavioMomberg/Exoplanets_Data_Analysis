## Primeira análise de dados sobre astronomia

Vamos analisar a base de dados Planetary Systems com intuito de aprender a realizar uma boa analise de dados.

O método utilizado:

1. Análisar a base de dados (ver suas colunas)
2. Descartar colunas que não ajudam na análise (Ex. colunas de ID)
3. Tratar os dados (dados duplicados, nulos, tipos)
4. Análise inicial
5. Análise aprofundada
6. Apresentar os resultados

Dataset URL = https://exoplanetarchive.ipac.caltech.edu/cgi-bin/TblView/nph-tblView?app=ExoTbls&config=PS  

```python
import pandas as pd

# ler a base de dados. 
df = pd.read_csv("exoplanet_dataset.csv", sep=",", comment="#", dtype={4: str , 5: str})

display(df)
```

```python
# selecionar algumas colunas para trabalhar
colunas = [
    "pl_name", "hostname", "discoverymethod", 
    "disc_year", "pl_orbper", "pl_rade",
    "st_mass", "st_rad", "st_teff", "sy_dist" 
]

df = df[colunas]
```

## Colunas selecionadas para a análise:

- (A massa do planeta não foi utilizada por cerca de 90% das linhas estarem constando NaN)

- pl_name -> nome do exoplaneta
- hostname -> nome da estrela
- discoverymethod -> método utilizado para descobrir o exoplaneta
- disc_year -> ano que o exoplaneta foi descoberto
- pl_orbper -> período orbital do exoplaneta (em dias)
- pl_rade -> raio do exoplaneta (considerando o raio da Terra. Ex. pl_rade = 1.0 significa que é o mesmo raio da Terra)
- st_mass -> massa da estrela (considera a massa solar)
- st_rad -> raio da estrela (considera o raio do Sol)
- st_teff -> temperatura da superfície da estrela (em Kelvin - K)
- sy_dist -> distância até a Terra (em persecs - pc, corresponde a ~3,26 anos luz)


```python
# tratar os dados (null, NaN, tipos errados)

display(df.isna().sum())

# remove valores nulos (quando não for em uma quantidade muito grande)
df = df.dropna()
display(df.info())

# guardar dados do tipo texto em uma variável separada
nomes_sem_duplicatas = df.drop_duplicates("pl_name")
colunas_nao_numericas = nomes_sem_duplicatas[["pl_name", "hostname", "discoverymethod"]]

# fazer a média dos valores numerícos (Ex. "pl_rade") agrupados pelo nome do planeta
df = df.groupby("pl_name").mean(numeric_only=True)

# unir ambas as variáveis (colunas_nao_numericas e df) para formar a tabela com os dados completos 
df = colunas_nao_numericas.merge(df, on="pl_name")
display(df)
```

```python
import plotly.express as px

# verificar a porcentagem de cada método utilizado para descobrimento de explanetas
display(df["discoverymethod"].value_counts(normalize=True).map("{:.2%}".format))

# construir um gráfico que demonstre a quantidade de vezes que cada método para descobrimento de exoplanetas foi utilizado
grafico = px.histogram(
    df, 
    x="discoverymethod", 
    title="Métodos de Descobrir Exoplanetas", 
    subtitle="Percentual de métodos de descobrir exoplanetas",
    text_auto=True,
)
grafico.show()
```

```python
# verificar a quantidade de exoplanetas que foram descobertos por ano
display(df["disc_year"].value_counts())

# construir um gráfico demonstrando a quantidade de exoplanetas que foram descobertos por ano e quais métodos foram utilizados
grafico = px.histogram(df, x="disc_year", color="discoverymethod")
grafico.show()
```

```python
# construindo um gráfico para mostrar estrelas filtradas
# filtro pela quantidade de massa da estrela no intervalo: 0.9 - 1.1 vezes a massa do Sol e com raio até duas vezes o raio do Sol
grafico = px.scatter(
    df[(df["st_mass"] > 0.9) & (df["st_mass"] < 1.1) & (df["st_rad"] <= 2)], 
    x="st_mass", 
    y="st_rad", 
    color="hostname",
    title="Estrelas similares ao Sol",
)
grafico.show()

# display para informar a quantidade de itens encontrados de acordo com o filtro aplicado no gráfico acima e outras informações como tipo de dado
display(df[(df["st_mass"] > 0.9) & (df["st_mass"] < 1.1) & (df["st_rad"] <= 2)].info())
```

```python
# varivel para armazenar linhas com estrelas semelhantes a algumas características físicas do Sol como massa e raio
sol_like = df[(df["st_mass"] > 0.9) & (df["st_mass"] < 1.1) & (df["st_rad"] <= 2)]

# filtrando as linhas restantes para armazenar aquelas que o planeta possui seu período orbital similar ao da Terra (até ~17% menor ou até ~91% maior) 
sol_like = sol_like[(sol_like["pl_orbper"] > 300) & (sol_like["pl_orbper"] < 700)]

# display de informações sobre os planetas encontrados de acordo com o filtro aplicado acima
display(sol_like.info())

# tabela organizando os planetas de acordo com seu período orbital (em dias) na ordem decrescente
display(sol_like.sort_values(by="pl_orbper", ascending=False))
```

```python
# função para inserir uma nova coluna chamada pl_class
def classificar_exoplanetas(pl_rade: float) -> str:
    if (pl_rade <= 0.8):
        return "mini_earth"
    if (pl_rade <= 1.25):
        return "earth-sized"
    if (pl_rade <= 1.75):
        return "super_earth"
    if (pl_rade <= 3):
        return "mini_neptune"
    if (pl_rade <= 6):
        return "neptune"
    else:
        return "gas_giant"
        
# receber valores para a nova coluna
sol_like["pl_class"] = sol_like["pl_rade"].apply(classificar_exoplanetas)

# display da tabela atualizada
display(sol_like)

# este gráfico mostra a distribuição dos planetas de acordo com seu raio
grafico = px.bar(
    sol_like.sort_values(by="pl_rade", ascending=False), 
    x="pl_rade", 
    y="pl_name", 
    color="pl_class",
    title="Exoplanetas de acordo com o raio",
    subtitle="Classificação dos exoplanetas considerando o raio"
)
grafico.show()
```

## Classificação dos exoplanetas de acordo com seu raio:

- Mini Terra (pl_rade > 0.5 e <= 0.8)

- Terra (pl_rade > 0.8 e <= 1.25)

- Super Terra (pl_rade > 1.25 e <= 1.75)

- Mini Netuno (pl_rade > 1.75 e <= 3)

- Netuno (pl_rade > 3 e <= 6)

- Gigante Gasoso (pl_rade > 6 e <= ~14.3)

(para informações a respeito do tipo de cada um desses planetas acesse o link do site da Nasa abaixo)

Fonte : https://science.nasa.gov/exoplanets/planet-types/#h-a-galaxy-of-stars-and-planets

## Conclusões

1698 estrelas similares ao Sol foram encontradas;

Essas estrelas foram encontradas através da seguinte filtragem:
 - A estrela candidata deve ter massa (st_mass) entre 0.9 e 1.1 (em massa solar, ou seja, 1 = 1 vez a massa do Sol);
 - A estrela deve também possuir um raio de até 2 vezes o raio do Sol.

12 exoplanetas foram encontrados com base no período orbital proxímo ao da Terra, porém todos eles sendo 
maiores que a Terra e possivelmente com suas respectivas massas também maiores (A massa não foi utilizada nessa análise pela baixa quantidade
de dados disponível neste campo na base de dados);

75% dos exoplanetas encontrados possuem ao menos 3 vezes o raio da Terra e desses 75%, temos 44% deles sendo no minímo 6 vezes maior que o nosso
planeta, portanto eles devem ser planetas gasosos;

Outra interessante observação que podemos realizar é que entre 2014 e 2016 houve uma explosão na descoberta de exoplanetas, representando mais de 
51% dos exoplanetas encontrados de acordo com as métricas aplicadas na filtragem dos dados utilizados no estudo
